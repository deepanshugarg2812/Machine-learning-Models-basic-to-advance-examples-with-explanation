#### Naive Bayes for feature classification with similaritiles to the digit data

```python
### Importing the data
import pandas as pds
uncleanedData = pds.read_csv('mushrooms.csv')

## Checking the type
print(type(uncleanedData))

### Stastical data analysis
uncleanedData.head(n=10)
```

```python
## Transforming the data to the numerical format
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
uncleanedData = uncleanedData.apply(le.fit_transform)
uncleanedData.head(n=10)
```

```python
### Extracting the data
Udata = uncleanedData.values
print(type(Udata),Udata.shape)

## Creating the training data
X = Udata[:,1:]
Y = Udata[:,0]

print(X.shape)
print(Y.shape)
```

```python
### Data model
import numpy as np

class NaiveBayes:
    def __init__(self,X,Y):
        """Initialize the data set"""
        self.X = X
        self.Y = Y
    
    def priorProb(self,label):
        """Return the prior prob"""
        Y = self.Y
        return np.sum(Y==label)/Y.shape[0] 
    
    def conditionalProb(self,indexOfFeature,featureVal,label):
        """Return the conditional prob p(a/b)"""
        X = self.X
        Y = self.Y
        
        x_filtered = X[Y==label]
        numerator = np.sum(x_filtered[:,indexOfFeature]==featureVal)
        denominator = np.sum(Y==label)
        return numerator/denominator
    
    def fitAndPredict(self,data):
        """Return the prediction based on the computation"""
        X = self.X
        Y = self.Y
        number_of_features = X.shape[1]
        labels = np.unique(Y)
        ## Storing the labels and there likelihood
        postiror = []
        for label in labels:
            likelihood = 1.0
            for i in range(number_of_features):
                likelihood *= self.conditionalProb(i,data[i],label)
            prior = self.priorProb(label)
            postiror.append(likelihood*prior)
        p = np.argmax(postiror)
        return p
```

```python
NB = NaiveBayes(X,Y)
print(NB.fitAndPredict(X[0]))
print(Y[0])
```

```python
print(NB.fitAndPredict(X[1]))
print(Y[1])
```

```python
y_Pred = []
for i in range(X.shape[0]):
    y_Pred.append(NB.fitAndPredict(X[i]))
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 

results = confusion_matrix(Y, y_Pred)
print(results)
print('Accuracy Score :',accuracy_score(Y, y_Pred))
```
